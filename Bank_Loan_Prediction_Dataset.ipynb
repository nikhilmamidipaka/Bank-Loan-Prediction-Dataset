````python
# Checking whether columns having null values
data.isnull().sum()
````

````python
# Removing rows having null values greater than or equal to 30%

emptyrow = data.isnull().sum(axis=1)
emptyrow = list(emptyrow[emptyrow.values >= 0.3 * len(data)].index)
data.drop(labels = emptyrow, axis = 0, inplace = True)
print(len(emptyrow))
````
